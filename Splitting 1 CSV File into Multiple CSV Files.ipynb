<h1><b>Splitting 1 CSV File into Multiple CSV Files (By Column)</h1></b>

```python
import pandas as pd
from tabulate import tabulate
```

```python
data_df = pd.read_csv(r"C:\Users\LENOVO\Documents\Personal Project\Data Analysis on DataCo Supply Chain Management System\DataCo Supply Chain Dataset.csv", encoding='latin-1') 
display(data_df)
```

```python
# List all columns in the original CSV file
all_columns = data_df.columns.tolist()
print("All Columns in the Original CSV File:")
for index, column in enumerate(all_columns, start=1):
    print(f"{index}. {column}")
```

```python
# Format the list in a table
columns_table = [{"Index": index, "Column Name": column} for index, column in enumerate(all_columns, start=1)]

# Print the table
print("All Columns in the Original CSV file:")
print(tabulate(columns_table, headers="keys", tablefmt="grid"))
```

<h2>Create Separate Dataframes for Each Table</h2>


<h2>Tables</h2>

1. Orders
2. OrderItems
3. Customers
4. Products
5. Categories
6. Department

```python
# Orders Table
orders_columns = ['Order Id', 'Order Customer Id', 'order date (DateOrders)', 'Order Status', 'Order Profit Per Order', 'Market', 'Order City', 'Order State', 'Order Country', 'Order Region', 'shipping date (DateOrders)']
orders_df = data_df[orders_columns]
display(orders_df)
```

```python
# Change the data type of columns Shipping Date into date
orders_df['shipping date (DateOrders)'] = pd.to_datetime(orders_df['shipping date (DateOrders)'], infer_datetime_format=True)
orders_df['Shipping Date'] = orders_df['shipping date (DateOrders)'].dt.strftime('%Y-%m-%d')

display(orders_df)
```

```python
# Change the data type of columns Order Date into date
orders_df['order date (DateOrders)'] = pd.to_datetime(orders_df['order date (DateOrders)'], infer_datetime_format=True)
orders_df['Order Date'] = orders_df['order date (DateOrders)'].dt.strftime('%Y-%m-%d')

display(orders_df)
```

```python
# OrderItems Table
order_items_columns = ['Order Item Id', 'Order Id', 'Product Card Id', 'Order Item Cardprod Id', 'Order Item Discount', 'Order Item Discount Rate', 'Order Item Product Price', 'Order Item Profit Ratio', 'Order Item Quantity', 'Order Item Total', 'Sales per customer', 'Order Profit Per Order']
order_items_df = data_df[order_items_columns]
display(order_items_df)
```

```python
# Customers Table
customers_columns = ['Customer Id', 'Customer Fname', 'Customer Lname', 'Customer Email', 'Customer Password', 'Customer Segment', 'Customer City', 'Customer State', 'Customer Country', 'Customer Street', 'Customer Zipcode', 'Latitude', 'Longitude']
customers_df = data_df[customers_columns]
display(customers_df)
```

```python
# Products Table
products_columns = ['Product Card Id', 'Product Category Id', 'Product Name', 'Product Description', 'Product Image', 'Product Price', 'Product Status']
products_df = data_df[products_columns]
display(products_df)
```

```python
# Categories Table
categories_columns = ['Category Id', 'Category Name']
categories_df = data_df[categories_columns]
display(categories_df)
```

```python
# Departments Table
departments_columns = ['Department Id', 'Department Name']
departments_df = data_df[departments_columns]
display(departments_df)
```

```python
# OrdersProcessing Table
ordersprocessing_columns = ['Order Id', 'Type', 'Days for shipping (real)', 'Days for shipment (scheduled)', 'Benefit per order', 'Shipping Mode', 'Delivery Status', 'Order Status', 'Late_delivery_risk']
ordersprocessing_df = data_df[ordersprocessing_columns]
display(ordersprocessing_df)                    
```

```python
# Replace 'output_path' with the desired output directory
output_path = 'C:\\Users\\LENOVO\\Documents\\Personal Project\\Data Analysis on DataCo Supply Chain Management System'

# Save each DataFrame to a separate CSV file
orders_df.to_csv('orders.csv', index=False)
order_items_df.to_csv('order_items.csv', index=False)
customers_df.to_csv('customers.csv', index=False)
products_df.to_csv('products.csv', index=False)
categories_df.to_csv('categories.csv', index=False)
departments_df.to_csv('departments.csv', index=False)
```
